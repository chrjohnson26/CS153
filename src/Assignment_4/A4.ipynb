```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Importing the images

```python
impath = 'imgs/cat.png'
cat = cv2.imread(impath)
cat = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
(cat_h,cat_w,cat_c) = cat.shape

impath = 'imgs/maple_leaf.png'
maple = cv2.imread(impath)
maple = cv2.cvtColor(maple, cv2.COLOR_BGR2RGB)
(maple_h,maple_w,maple_c) = maple.shape

impath = 'imgs/pikachu.png'
pikachu = cv2.imread(impath)
pikachu = cv2.cvtColor(pikachu, cv2.COLOR_BGR2RGB)
(pikachu_h,pikachu_w,pikachu_c) = pikachu.shape

impath = 'imgs/encode_A.png'
ea = cv2.imread(impath)
ea = cv2.cvtColor(ea, cv2.COLOR_BGR2RGB)
(ea_h,ea_w,ea_c) = ea.shape

impath = 'imgs/encode_B.png'
eb = cv2.imread(impath)
eb = cv2.cvtColor(eb, cv2.COLOR_BGR2RGB)
(eb_h,eb_w,eb_c) = eb.shape

impath = 'imgs/encode_C.png'
ec = cv2.imread(impath)
ec = cv2.cvtColor(ec, cv2.COLOR_BGR2RGB)
(ec_h,ec_w,ec_c) = ec.shape
```

```python
plt.subplot(1,3,1)
plt.imshow(cat)
plt.subplot(1,3,2)
plt.imshow(maple)
plt.subplot(1,3,3)
plt.imshow(pikachu)
plt.show()
```
